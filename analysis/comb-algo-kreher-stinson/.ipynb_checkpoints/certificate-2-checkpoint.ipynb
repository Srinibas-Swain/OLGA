# In theory

Let G be a permutation group on $\mathcal{X} = \Brace{1, \dots, n-1}$ and set
\begin{array}{lll}
G_0 &= \{g \in G   &\mid ~g(0) = 0\} \\
G_1 &= \{g \in G_0 &\mid ~g(1) = 1\} \\
& \vdots  \\
G_{n-1} &= \{g \in G_{n-2} &\mid ~g(n-1) = n-1\} = \Brace{\I}\\
\end{array}

$$
\text{also, } G \supseteq G_0 \supseteq \dots \supseteq G_{n-1} = \Brace{\I}
$$

\begin{array}{lll}
\text{Let }       & \mathbf{orb}(0) &= \Brace{g(0) \mid g \in G} \\
\text{Let }       & n_0             &= \Mod{\mathbf{orb}} \\
\text{Then }      & \mathbf{orb}(0) &= \Brace{x_{0,1}, x_{0,2}, \dots, x_{0, n_0}}\\
\text{Let }       & \mathcal{U}_0   &= \Brace{h_{0,1}, h_{0,2}, \dots, h_{0, n_0}}\\
\text{such that } & h_{0,i}(0)      &= x_{0,i}\\
\end{array}

That is, find one permutation in G which maps zero to an orbital element 

## Theorem 6.6

Let $G, \Ua_0, G_0$ be as defined above. Then $\Ua_0$ is a left traversal of $G_0$ in G.  
That is 
$$
\Ua_0 G_0 = G
$$

## Theorem 6.7

Let $G, \Ua_i, G_i$ be as defined above. Then $\Ua_i$ is a left traversal of $G_i$ in G.  
That is 
$$
\Ua_i G_i = G_{i-1}
$$

Thus, it is clear as a day that, for every $g \in G$,
$$
g = 
\underbrace{h_{0, i_0}}_{\Ua_0}~
\overbrace{h_{1, i_1}}^{\Ua_1}~
\underbrace{h_{2, i_2}}_{\Ua_2}~
\cdots~
\underbrace{h_{n-1, i_{n-1}}}_{\Ua_{n-1}}
=
\prod_{j=0}^{n-1} h_{j, j_i}
$$

## Checking for membership

To check if $g \in G$

1. Step 0
  1. Let $x = g(0)$
  1. $\algoif \exists h_0 \in \Ua_0 ~|~ h_0(0) = x$
    1. $\commentblock{
        (\inv{h_0} g)(0) = 0 \\
        \text{that is } \inv{h_0} g \in G_0 \\
        f_0 \gets h_0 \\
        g_0 \gets \inv{h_0} g = \inv{f_0} g
        }{$\ds g = \prod_{i=0}^{n-1} h_i $;    Hence this amounts to peeling the rhs one at a time}
       $

1. Step i
  1. Let $x = g_{i-1}(i)$
  1. $\algoif \exists h_i \in \Ua_i ~|~ h_i(i) = x$
    1. $\commentblock{
        (\inv{h_i} g_{i-1})(i) = i\\
        \text{that is } \inv{h_i} g_{i-1} \in G_i \\
        f_i \gets f_{i-1} h_i \\
        g_i \gets \inv{h_i} g_{i-1} ~~ ( = \inv{f_i} g)
        }{$\ds g = \prod_{i=0}^{n-1} h_i $;    Hence this amounts to peeling the rhs one at a time}
       $

## Algorithm 6.8: test

$\algofun{test}(g, \bar{G} = \Bracket{\Ua_0, \cdots, \Ua_{n-1}})$

* $n \gets g.num\_vertices$
* $\algofor i \in [0, n-1)$
  * $x \gets g[i]$
  * $\algoif \exists h \in \Ua_i ~|~ h(i) = x$
    * $\commentblock{
        \pi_2 &\gets \inv{h} \\
        \pi_3 &\gets \pi_2 \times g \\
        g &\gets \pi_3
    }{Since $\ds g = \prod_{i=0}^{n-1} h_i $;    Hence this amounts to peeling the rhs one at a time}$
  * $\algoelse \algoreturn i$
* $\algoreturn n$

## refine

Algorithm 7.5

$\algofun{refine}(graph, A)$

* $\commentblock{B &= A.clone() \\ S &= B.clone().reverse()}{create B, S}$


* $\commentblock{
    n &= graph.num\_vertices  \\
    U &= Block.with\_range(0, n) \commentgreen{# from 0 till n-1}  \\
    N &= B.size() \commentgreen{# size of partition B}
   }{create U}
  $


* $\algowhile(N>0)$:  
  * $\left. \begin{matrix}
              N = N - 1 \\
              ~~T = S.pop()
            \end{matrix} \right\}
      \commentgreen{# consider the last element of S}$
  * $\algoif T \subset U$
    * U = U \ T $\commentgreen{# rm duplicates}$
    * j = 0
    
    * $\algowhile j < |B| \land |B| < n)$:
      * **if** $|B| \ne 1$
        * $\commentblock{\algofun{split\_and\_update}(graph, B, S, T, U, j, N) \\ j++}{Split and Update each block of B}$

  
* return B


### split-and-update

$\algofun{split\_and\_update}(graph, B, S, T, U, \&j, \&N)$

* L = Partition(graph.num_vertices) $\commentgreen{# Create a partition with n empty blocks}$ 


* $\algofor u \in B[j]$
  * $\commentblock{
       h &= | T \cap N_{\text{graph}}(u) | \\
       L[h] &= L[h] \cup \Paran{u}
      }{Update L}
    $


* $\commentblock{
     m = 0 \\
     \algofor h \in [0, n) \\
     ~~~~ \algoif L[h] \ne \phi \\
     ~~~~ ~~~~~ m++
   }{Find no. of non-empty blocks in L}
  $


* $\algoreturn \algoif m \le 1$ $\commentgreen{Return if B is already refined}$


* $\commentblock{
        \algofor h \in [|B|-1, j+1]: \\
        ~~ ~~ ~~ B[m-1+h] = B[h]
     }{Create a gap of size (m-1) after $j^{th}$ position}
    $


* k = 0
* S.allocate(N+m)
* $\algofor \algovar{block} \in L.blocks$:
  * $\algocontinue \algoif \algovar{block}$.empty?
  
  * $\commentblock{
         B[j+k] &= S[N+m-1+k] = \algovar{block} \\
         U &= U \cup \algovar{block} \\
         k ++
       }{Update B, S, U; update S from the last block}
    $
    
    
* $\commentblock{j &+= m-1 \\ N &+= m}{Update j, N}$

## canon2 

Algorithm 7.9  **canon2** => **canon2**(graph, partition p, partition $\mu$)

* $Q \gets \algofun{refine} (\algovar{graph}, \algovar{P})$
* $\algovar{L} = $Index of first block of Q with $|Q[L]| \gt 1$
* $\algovar{result}$ = "Better" 
* $ \algovar{result} = \algofun{compare}(Q, \mu) \algoif \mu \ne \phi$


* $\algoif |Q| == n$
  * $\algofun{Canon2Discrete}(\algovar{Q, result})$
  
* $\algoif \algovar{result} \ne$ "Worse"
  * $\algofun{Canon2NotWorse(Q, L)}$

### Canon2Discrete
$\commentgreen{If Q is refined, set $\mu$ if any of these conditions are met}$  
$\commentgreen{$\mu == \phi$ or result is "Better"}$  
$ \commentgreen{if result is "Equal", update the automorphism}$

$\algofun{Canon2Discrete}$(Q, n, $\mu$)

* $\algoif \algovar{\mu} == \phi$
  * $\algovar{\mu}$ = Q.clone()
* $\algoelsif \algovar{result}$ == "Better"
  * $\algovar{\mu}$ = Q.clone()
* $\algoelsif \algovar{result}$ == "Equal"
  * $\algocomment{Update the automorphisms}$
  * $\algofor i \in [0, n)$
    * $\algovar{\pi_2}[\algovar{Q}[i]] = \algovar{\mu}[i]$
    * $\algovar{sch}.\algofun{enter2}(\algovar{\pi_2})$ 
* $\algoreturn \algovar{\mu}$

#### enter2

$sch.\algofun{enter2}(\algovar{g})$

* $\algovar{N} \gets |g|$
* $\algovar{i} \gets \algofun{test2}(sch, g)$
* $\algoif i == N$
  * $\algoreturn$
* $\algoelse$
  * $\mathcal{U}_i \gets \mathcal{U}_i \cup g$
  
* $\algofor j \in [0, i]$
  * $\algofor \algovar{h} \in \mathcal{U}_j$
    * $\algovar{f} \gets \algovar{g} \times \algovar{h}$
    * $\algofun{enter2}(sch, \algovar{f})$

#### test2

$sch.\algofun{test2}(g)$

* $N \gets |g|$
* $\algofor i \in [0, N)$
  * $x \gets g[\beta[i]]$
  * $\algoif ! \exists h \in \mathcal{U}_i ~~s.t.~~ h[\beta[i]] == x$
    * $\algoreturn i$
  * $\algovar{g} \gets \inv{h} \times \algovar{g}$
* $\algoreturn N$

### Canon2NotWorse

$\commentgreen{if result is not "Worse"}$  
$\commentgreen{find the best split of Q[L] into a {singleton, rest of Q[L]} }$

$\algofun{Canon2NotWorse}(Q, L)$
* $\algovar{C} = Q[L].clone()$
* $\algowhile(|C|>0)$:
    * $\algovar{u} = \algovar{C}.pop()$
    * $R = \Brace{Q[:L-1], \algofocus{u, Q[L] $\setminus$ u}, Q[L+1:]} \commentgreen{Create two spaces at L, L+1 in Q}$
    * $\algovar{\mu} = \algofun{canon2}(graph, R, \mu)$
    * $\algovar{\beta'} = \Brace{R[0],\cdots,R[L]}$
    * $\algovar{j} = L$
    * $\commentblock{
       \algofor \algovar{y} \notin \Brace{\beta'[0],\cdots,\beta'[L]} \\
       ~~ ~~ ~~ \algovar{\beta'}[++j] = y
      }{fill $\beta'[L:]$ with vertices not in $\beta'[:L]$}$
    * $\algovar{sch} \gets \algovar{sch}.\algofun{change\_base}(\algovar{\beta'})$
    * $\algofor g \in \algovar{sch}[L]$
      * $C = C \setminus g[u]$

#### change-base

$\algofun{change\_base}(sch, \beta')$

* $N \gets |\beta'|$
* $\algovar{sch'} = SchreierSims(N, \beta')$
* $\algofor j \in [0, N)$
  * $\algofor g \in \mathcal{U}_j$
    * $\algovar{sch'}.\algofun{enter2}(g)$
* $\algoreturn \algovar{sch'}$
